Implement different morphological algorithm.

```python
#erosion
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image (make sure it's a binary image)
image = cv2.imread('/content/binary-image.png')

# Define a kernel (structuring element)
kernel = np.ones((5, 5), np.uint8)

# Perform erosion
erosion_result = cv2.erode(image, kernel, iterations=1)

# Display the original and eroded images
cv2_imshow( image)
cv2_imshow( erosion_result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#dialation

# Define a kernel (structuring element)
kernel = np.ones((5, 5), np.uint8)

# Perform dilation
dilation_result = cv2.dilate(image, kernel, iterations=1)

# Display the original and dilated images
cv2_imshow( image)
cv2_imshow( dilation_result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#opening


# Define a kernel (structuring element)
kernel = np.ones((5, 5), np.uint8)

# Perform opening
opening_result = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

# Display the original and opening results
cv2_imshow( image)
cv2_imshow(opening_result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#closing

# Define a kernel (structuring element)
kernel = np.ones((5, 5), np.uint8)

# Perform closing
closing_result = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

# Display the original and closing results
cv2_imshow( image)
cv2_imshow( closing_result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
