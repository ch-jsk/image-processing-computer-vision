```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
import numpy as np
```

```python
def average_filter(image,kernal_size):
  return cv2.blur(image,(kernal_size,kernal_size))
```

```python
image_path="/content/hawk.png"
image=cv2.imread(image_path)
```

average filter


```python
kernal_size=5
smoothed_image=average_filter(image,kernal_size)
```

```python
cv2_imshow(image)
cv2_imshow(smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

median filter


```python
def median_filter(image, kernel_size):
    return cv2.medianBlur(image, kernel_size)

# Load the image
image_path ="/content/hawk.png"
image = cv2.imread(image_path)

# Apply the median filter with a kernel size of 5x5
kernel_size = 5
median_image = median_filter(image, kernel_size)

# Display the original and filtered images
cv2_imshow( image)
cv2_imshow( median_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

max filter

```python
def max_filter(image, kernel_size):
    return cv2.dilate(image, np.ones((kernel_size, kernel_size), np.uint8))

image_path = "/content/hawk.png"
image = cv2.imread(image_path)

kernel_size = 5
max_image = max_filter(image, kernel_size)

# Display the original and filtered images
cv2_imshow(image)
cv2_imshow( max_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

mean filter


```python

def mean_filter(image, kernel_size):
    return cv2.blur(image, (kernel_size, kernel_size))

# Load the image
image_path = "/content/hawk.png"
image = cv2.imread(image_path)

# Apply the mean filter with a kernel size of 5x5
kernel_size = 5
mean_image = mean_filter(image, kernel_size)

# Display the original and filtered images
cv2_imshow( image)
cv2_imshow( mean_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
