Read an image and apply (1) Gaussian 3x3 mask for burring (2) High pass filter
mask with different masks

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path="/content/112.jfif"
img=cv2.imread(image_path)
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis('off')
plt.title("original")
plt.show()
```

gaussian blur


```python
blurred_img=cv2.GaussianBlur(img_rgb,(3,3),0)
```

```python
plt.imshow(blurred_img)
plt.axis('off')
plt.title("Blurred image (Gaussian 3x3)")
plt.show()
```

high pass filter mask

```python
laplacian = cv2.Laplacian(img_rgb,cv2.CV_64F)
laplacian=np.uint8(np.absolute(laplacian))
```

```python
plt.imshow(laplacian,cmap='gray')
plt.axis('off')
plt.title("high Pass Filter(Laplacian)")
plt.show()
```
