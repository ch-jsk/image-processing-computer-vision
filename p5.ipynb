```python
#A) Implement Gray level slicing (intensity level slicing) in to read cameraman image
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

```

```python
image=cv2.imread('cmi.png',cv2.IMREAD_GRAYSCALE)
```

```python
lower_level=100
upper_level=200
```

```python
output = np.copy(image)
output[(image>=lower_level)&(image<=upper_level)]=255
output[(image<lower_level)|(image>upper_level)]=0
```

```python
fig,axs=plt.subplots(1,2,figsize=(10,5))
axs[0].imshow(image,cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')
axs[1].imshow(output,cmap='gray')
axs[1].set_title('Gray Level Slicing')
axs[1].axis('off')
plt.show()
```

```python
#(B)  Read an 8bit image and to see the effect of each bit on the image.
```

```python
height,width=image.shape
```

```python
fig, axs = plt.subplots(4, 2, figsize=(8, 16))
axs = axs.flatten()

for bit in range(8):
    mask = 1 << bit
    bit_image = np.bitwise_and(image, mask)
    bit_image = (bit_image / mask) * 255
    bit_image = bit_image.astype(np.uint8)

    axs[bit].imshow(bit_image, cmap='gray')
    axs[bit].set_title('Bit {}'.format(bit))
    axs[bit].axis('off')

plt.tight_layout()
plt.show()
```
