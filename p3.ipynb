Implement different Interpolation techniques.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = "moon.jpg"
image = cv2.imread(image_path)
```

```python
scale_percent=50
```

```python
nearest_neighbor=cv2.resize(image,None, fx=scale_percent/100,fy=scale_percent/100,interpolation=cv2.INTER_NEAREST)
```

```python
bilinear = cv2.resize(image,None,fx=scale_percent/100,fy=scale_percent/100, interpolation=cv2.INTER_LINEAR)
```

```python
bicubic = cv2.resize(image,None,fx=scale_percent/100,fy=scale_percent/100, interpolation=cv2.INTER_CUBIC)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.subplot(2,2,2)
plt.imshow(cv2.cvtColor(nearest_neighbor,cv2.COLOR_BGR2RGB))
plt.title('Nearest Neighbor')
plt.subplot(2,2,3)
plt.imshow(cv2.cvtColor(bilinear,cv2.COLOR_BGR2RGB))
plt.title('Bilinear')
plt.subplot(2,2,4)
plt.imshow(cv2.cvtColor(bicubic,cv2.COLOR_BGR2RGB))
plt.title('Bicubic')
```
