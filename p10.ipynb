```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('img123.jpg', cv2.IMREAD_GRAYSCALE)

# Define the Sobel, Prewitt, and Roberts masks
sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewitt_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])

roberts_x = np.array([[1, 0], [0, -1]])
roberts_y = np.array([[0, 1], [-1, 0]])

# Apply the masks to the image
sobel_x_filtered = cv2.filter2D(image, -1, sobel_x)
sobel_y_filtered = cv2.filter2D(image, -1, sobel_y)
sobel_combined = np.sqrt(sobel_x_filtered**2 + sobel_y_filtered**2)

prewitt_x_filtered = cv2.filter2D(image, -1, prewitt_x)
prewitt_y_filtered = cv2.filter2D(image, -1, prewitt_y)
prewitt_combined = np.sqrt(prewitt_x_filtered**2 + prewitt_y_filtered**2)

roberts_x_filtered = cv2.filter2D(image, -1, roberts_x)
roberts_y_filtered = cv2.filter2D(image, -1, roberts_y)
roberts_combined = np.sqrt(roberts_x_filtered**2 + roberts_y_filtered**2)

# Plot the original and filtered images
plt.figure(figsize=(12, 8))
plt.subplot(2, 4, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(2, 4, 2), plt.imshow(sobel_x_filtered, cmap='gray'), plt.title('Sobel X')
plt.subplot(2, 4, 3), plt.imshow(sobel_y_filtered, cmap='gray'), plt.title('Sobel Y')
plt.subplot(2, 4, 4), plt.imshow(sobel_combined, cmap='gray'), plt.title('Sobel Combined')

plt.subplot(2, 4, 5), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(2, 4, 6), plt.imshow(prewitt_x_filtered, cmap='gray'), plt.title('Prewitt X')
plt.subplot(2, 4, 7), plt.imshow(prewitt_y_filtered, cmap='gray'), plt.title('Prewitt Y')
plt.subplot(2, 4, 8), plt.imshow(prewitt_combined, cmap='gray'), plt.title('Prewitt Combined')

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(1, 3, 2), plt.imshow(roberts_x_filtered, cmap='gray'), plt.title('Roberts X')
plt.subplot(1, 3, 3), plt.imshow(roberts_y_filtered, cmap='gray'), plt.title('Roberts Y')

plt.show()
```
