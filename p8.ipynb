Read an image and apply  (1) Laplacian operator with center value positive and
negative (2) High boost filtering.

```python
from matplotlib import *
from pylab import *
import cv2
import numpy as np
```

```python
from google.colab.patches import cv2_imshow
```

```python
image_path = "/content/hawk.png"
img_rgb = cv2.imread(image_path)
cv2_imshow(img_rgb)
```

```python
laplacian_pos=cv2.Laplacian(img_rgb,cv2.CV_64F)
laplacian_pos=np.uint8(np.clip(laplacian_pos+128,0,255))
laplacian_neg=cv2.Laplacian(img_rgb,cv2.CV_64F)
laplacian_neg=np.uint8(np.clip(laplacian_neg-128,0,255))
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(laplacian_pos, cmap='gray')
plt.axis('off')
plt.title("Laplacian(Positive Center)")
plt.subplot(1,2,2)
plt.imshow(laplacian_neg, cmap='gray')
plt.axis('off')
plt.title("Laplacian(Negative Center)")
```

```python
k=1.5
mask=np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
high_boost_img= cv2.filter2D(img_rgb,-1,mask)
```

```python
plt.imshow(high_boost_img)
plt.axis('off')
plt.title("High Boost Filter")
plt.show()
```
